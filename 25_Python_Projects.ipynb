# **Guess the Number Game Python Project (computer)**

```python
import random

def computer_guess():
    low = 1
    high = 10
    feedback = ''

    print("Think of a number between 1 to 10, and I (computer) will try to guess it!")

    while feedback != 'c':
        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low  # or high, both are same here

        feedback = input(f"Is {guess} too High (H), too Low (L), or Correct (C)? ").lower()

        if feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1

    print(f"Yay! I guessed your number: {guess} ğŸ‰")

# Function call to run the game
computer_guess()
```

# **Guess the Number Game Python Project (user)**


```python
import random

print("Welcome to the Guess the Number Game!")
print("Let's Play Guess the Number Game!")

choices = ['easy', 'medium', 'hard']
user_score = computer_score = 0

while True:
    user_input = input("Choose difficulty (easy, medium, hard) or type 'quit' to exit: ").lower()

    if user_input == 'quit':
        print("Thanks for playing!")
        break

    if user_input not in choices:
        print("Invalid choice. Please choose easy, medium, or hard.")
        continue

    if user_input == 'easy':
        number = random.randint(1, 10)
        attempts = 5
    elif user_input == 'medium':
        number = random.randint(1, 20)
        attempts = 4
    elif user_input == 'hard':
        number = random.randint(1, 50)
        attempts = 3

    print(f"Guess the number! You have {attempts} attempts.")

    for attempt in range(attempts):
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("Please enter a valid number.")
            continue

        if guess == number:
            print("Correct! You win this round.")
            user_score += 1
            break
        elif guess < number:
            print("Too low!")
        else:
            print("Too high!")

    else:
        print(f"You ran out of attempts. The number was {number}.")
        computer_score += 1

    print(f"Score -> You: {user_score} | Computer: {computer_score}\n")
```

```python
import random

stages = [
    """
        ___________
       |           |
       |
       |
       |
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |
       |
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |           |
       |
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |          /|
       |
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |          /|\\
       |
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |          /|\\
       |          /
       |
       |
     __|___
    """,
    """
        ___________
       |           |
       |           O
       |          /|\\
       |          / \\
       |
       |
     __|___
    """
]


words = ["apple", "banana", "grape", "orange", "kiwi", "mango", "peach"]


chosen_word = random.choice(words)

word_display = ['_'] * len(chosen_word)


attempts = 6


while attempts > 0 and '_' in word_display:
    print("\nCurrent word: " + ' '.join(word_display))
    print(f"Remaining attempts: {attempts}")
    print(stages[6 - attempts])

    guess = input("Guess a letter: ").lower()


    if len(guess) != 1 or not guess.isalpha():
        print("Please enter a valid single letter.")
        continue


    if guess in chosen_word:
        print(f"Good guess! {guess} is in the word.")
        for i in range(len(chosen_word)):
            if chosen_word[i] == guess:
                word_display[i] = guess
    else:
        print(f"Oops! {guess} is not in the word.")
        attempts -= 1

if '_' not in word_display:
    print("\nCongratulations! You've guessed the word:", chosen_word)
else:
    print("\nGame over! The word was:", chosen_word)
```

```python
import time

def main(seconds):
    if seconds >= 0:
        mins = seconds // 60
        secs = seconds % 60
        print(f"{mins:02d}:{secs:02d}", end="\r")
        time.sleep(1)
        main(seconds - 1)
    else:
        print("\nTime's up!")

if __name__ == "__main__":
    main(10)
```

```python
import random
import string

def generate_password(length):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password


password_length = input("Enter the desired length of the password: ")

try:
    password_length = int(password_length)
    if password_length <= 0:
        print("Password length must be a positive integer.")
    else:
        password = generate_password(password_length)
        print("Generated Password:", password)
except ValueError:
    print("Please enter a valid integer.")
```

```python
import streamlit as st

st.set_page_config(page_title="BMI Calculator", page_icon="âš–ï¸")

st.title("âš–ï¸ BMI Calculator")

# User inputs
weight = st.number_input("Enter your weight in kg:", min_value=0.0, value=60.0, step=0.5)
height = st.number_input("Enter your height in cm:", min_value=0.0, value=170.0, step=0.5)

# Convert cm to meters
final_height = height / 100

if st.button("Calculate BMI"):
    if final_height > 0:
        bmi = weight / (final_height ** 2)
        st.success(f"âœ… Your BMI is: **{bmi:.2f}**")

        # BMI Category
        if bmi < 18.5:
            st.warning("âš ï¸ You are **underweight**. It's recommended to gain some weight.")
        elif 18.5 <= bmi < 24.9:
            st.info("ğŸ‰ You have a **normal** weight. Good job!")
        elif 25 <= bmi < 29.9:
            st.warning("âš ï¸ You are **overweight**. Try to stay active and eat healthy.")
        else:
            st.error("â— You are **obese**. It's highly recommended to consult a healthcare provider.")
    else:
        st.error("ğŸš« Height must be greater than 0.")

# Optional: BMI Range Info Table
with st.expander("â„¹ï¸ BMI Categories"):
    st.markdown("""
    | BMI Range       | Category       |
    |------------------|----------------|
    | Less than 18.5   | Underweight    |
    | 18.5 - 24.9      | Normal weight  |
    | 25 - 29.9        | Overweight     |
    | 30 and above     | Obese          |
    """)
```
